```python
# Bi Variate Analysis
```

```python
plt.figure(figsize=(16,4))
plt.scatter(data=df,x='Science',y='SAT')
plt.xlabel("Science")
plt.ylabel("SAT")
plt.title('Relation between Science and Sat')
plt.show()
```

```python
plt.figure(figsize=(16,4))
plt.scatter(data=df,x='General_knowledge',y='SAT')
plt.xlabel("General_knowledge")
plt.ylabel("SAT")
plt.title('Relation between General_knowledge and Sat')
plt.show()
```

```python
plt.figure(figsize=(25,4))
sns.jointplot(data=df,x='General_knowledge',y='SAT')
plt.xlabel('GK')
plt.xlabel('SAT')
plt.title('Relation between General_knowledge and Sat')
```

```python
plt.figure(figsize=(25,4))
sns.jointplot(data=df,x='General_knowledge',y='SAT',kind='hex')
plt.xlabel('GK')
plt.xlabel('SAT')
plt.title('Relation between General_knowledge and Sat')
```
