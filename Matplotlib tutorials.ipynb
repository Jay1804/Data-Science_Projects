
## Line Colors and Styles

```python
plt.plot(x, np.sin(x - 0), color='blue')        # color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk) 
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1 
plt.plot(x, np.sin(x - 5), color='chartreuse'); # HTML color names
plt.show()
```

```python
plt.plot(x, x + 0, linestyle='solid')  
plt.plot(x, x + 1, linestyle='dashed')   
plt.plot(x, x + 2, linestyle='dashdot')   
plt.plot(x, x + 3, linestyle='dotted')
```

```python
# Short codes for linestyles
plt.plot(x, x + 4, linestyle='-')  # solid  
plt.plot(x, x + 5, linestyle='--') # dashed  
plt.plot(x, x + 6, linestyle='-.') # dashdot   
plt.plot(x, x + 7, linestyle=':')  # dotted
```

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black     
plt.plot(x, x + 3, ':r')  # dotted red
```

- plt.plot()  ax.plot()

- plt.xlabel()  ax.set_xlabel() 
- plt.ylabel()  ax.set_ylabel() 
- plt.xlim()  ax.set_xlim()

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
plt.bar(company,revenue,label='revenue')
plt.bar(company,profit,label='revenue')
plt.legend()
plt.xlabel("Stocks")
plt.ylabel("revenue")
```

```python
#plotting revenue and profit side by side
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
xpos=np.arange(len(company))
plt.xticks(xpos,company)
plt.bar(xpos-0.2,revenue,width=0.3,label='revenue')
plt.bar(xpos+0.2,profit,color='blue',width=0.3,label='profit')
plt.legend()
plt.xlabel("Stocks")
plt.ylabel("revenue and profit")
```

## Scatter Plot
- plt.scatter has the capibilty to render a differnt size/ color for each point.
- plt.plot points are always essentially clones of each other, i.e., determination of appearence of points is done only once for the entire set of data
- This difference leads to vast diffreence in performance for large datasets

```python
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [5, 2, 4, 2, 1, 4, 5, 2]
plt.scatter(x, y, label = 'cat', color = 'k', s=130 , marker = "^")  # s is marker size

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot")

plt.legend()
plt.show()
```

```python
# Scatter plot using plt.plot()
x = np.linspace(0, 10, 50)
y = np.sin(x)
plt.plot(x, y, 'o', color = 'blue')
```

```python
y = np.linspace(0,10,50)
dx = 0.8
x =np.sin(y) + dx * np.random.randn(50)
plt.errorbar(x, y, xerr=dx, fmt='o', color='black', 
             ecolor='lightgray', elinewidth=3, capsize=0)
```

## Area Graph/ Stack Plot

```python
days = [1, 2, 3, 4, 5]

sleeping = [7, 8, 6, 11, 7]
eating = [2, 3, 4, 3, 2]
coding = [7, 8, 7, 2, 2]
playing = [8, 5, 7, 8, 13]

plt.plot([], [], color = 'm', label = 'Sleeping', linewidth = 5)
plt.plot([], [], color = 'r', label = 'eating', linewidth = 5)
plt.plot([], [], color = 'k', label = 'coding', linewidth = 5)
plt.plot([], [], color = 'c', label = 'playing', linewidth = 5)

plt.stackplot(days, sleeping, eating, coding, playing, colors = ['m', 'r', 'k', 'c'])

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Stack Plot")

plt.legend()
plt.show()
```

## Pie Chart
- for percentages or a fraction of a whole

```python
fractions = [7, 2, 2, 13]
activities = ['sleeping', 'eating', 'coding', 'playing']
cols = ['c', 'm', 'r', 'b']

plt.pie(fractions , labels = activities, colors = cols, startangle = 90, shadow = True, explode = (0, 0.1, 0, 0), autopct = '%1.1f%%')

plt.title('Pie Plot')
plt.show()
```

How to change size or resize figure

```python
fig = plt.figure()
axis = fig.add_axes([0.3 , 0.8 , 0.8 , 0.2]) # Measurement of canvas: left, right, width, height
axis.plot(x,y,'r')
```

```python
fig = plt.figure()
axis = fig.add_axes([0.8 , 0.8 , 0.8 , 0.9]) # Measurement of canvas: left, right, width, height
axis.plot(x,y,'r')
```

```python
fig , ax = plt.subplots(nrows = 2, ncols = 1)
for axis in ax:
    axis.plot(x,y ,'r')
```

```python
#subplot by hand
fig = plt.figure()
axis1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axis2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axis1.plot(x, y, 'r')
axis2.plot(y, x, 'b')
```
